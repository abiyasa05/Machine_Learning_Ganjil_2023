```python
!pip install pytesseract
!sudo apt-get install tesseract-ocr-ind

import cv2
import matplotlib.pyplot as plt
import pytesseract

# Baca citra KTP
wa_img = cv2.imread('/content/drive/MyDrive/Machine Learning/Kuis 2/kartu.png')
wa_gray = cv2.cvtColor(wa_img, cv2.COLOR_BGR2GRAY)

# Terapkan global thresholding pada citra KTP
global_threshold, wa_thresholded = cv2.threshold(wa_gray, 127, 255, cv2.THRESH_BINARY)

# Tampilkan citra hasil thresholding
plt.imshow(wa_thresholded, cmap='gray')
plt.show()

# Ekstrak teks dari citra hasil thresholding
extracted_text = pytesseract.image_to_string(wa_thresholded, lang="ind")

# Tampilkan teks yang diekstrak
print("\nHasil Ekstrak Teks\n\n"+extracted_text)
```
